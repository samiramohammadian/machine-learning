```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
```

```python
x_train = np.random.uniform(0,20, 200)
y_train = x_train+ np.random.uniform(-2 , 2 , 200) * np.random.uniform(1, 2, 200)
```

```python
x = x_train.reshape((200,1))
y = y_train.reshape((200 , 1))
```

```python
plt.scatter(x, y, marker='h')
```

### random line

```python
plt.scatter(x ,y , c = 'blue' , marker='h' , alpha = 0.5 , label = 'Data')
y_pred = x * 1 

plt.plot(x,y_pred , c = 'red' , label='random line 1')

y_pred = x * 0.75 

plt.plot(x,y_pred , c = 'green', label='random line 1')

y_pred = x * 1.5 

plt.plot(x,y_pred , c = 'hotpink', label='random line 1')

plt.legend()
plt.show()
```

### LLS line

```python
m = np.matmul(np.linalg.inv(np.matmul(x.T, x)), np.matmul(x.T, y))
m = m[0,0]
```

```python
y_pred_cor = x* m
plt.scatter(x ,y , c = 'blue' , marker='h' , alpha = 0.5 ,label = 'Data')
plt.plot(x,y_pred_cor , c = 'red' , label='TRue line')
plt.legend()
plt.show()
```
