```python
X = boston[[1,6]]
X.head()
```

```python
Y = boston[[12]]
Y.head()
```

```python
X = X.to_numpy()
Y = Y.to_numpy()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0],X[:,1],Y , c= 'hotpink' )
ax.set_xlabel("zn")
ax.set_ylabel("age")
ax.set_zlabel("medv")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_train[:,0],X_train[:,1],y_train , c= 'hotpink' )
ax.set_xlabel("zn")
ax.set_ylabel("age")
ax.set_zlabel("medv")
plt.show()
```

```python
m = np.matmul(np.linalg.inv(np.matmul(X.T, X)), np.matmul(X.T, Y))
m
```

```python
fig = plt.figure()
ax = fig.add_subplot( projection='3d')

x_range = np.arange(X[:,0].min(), X[:,0].max())
y_range = np.arange(X[:,1].min(), X[:,1].max())
x, y = np.meshgrid(x_range, y_range)
z = x * m[0] + y * m[1]

ax.plot_surface(x, y, z)
ax.set_xlabel("zn")
ax.set_ylabel("age")
ax.set_zlabel("medv")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')
ax.plot_surface(x, y, z, alpha = 0.4 )
ax.scatter(X[:,0],X[:,1],Y, c='red')
ax.set_xlabel("zn")
ax.set_ylabel("age")
ax.set_zlabel("medv")
plt.show()
```
