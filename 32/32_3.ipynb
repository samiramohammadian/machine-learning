```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# width & length & weight  melon 
# width & length & weight  balloon 
N = 250
std = 1.2

melons = pd.DataFrame({
    'width': np.random.normal(8, std, N),
    'height': np.random.normal(14, std, N),
    'weight': np.random.normal(28, std, N)}
)

ballons = pd.DataFrame({
    'width': np.random.normal(15, std, N),
    'height': np.random.normal(17, std, N),
    'weight': np.random.normal(5, std, N)}
)
```

```python
figre = plt.figure()

ax = figre.add_subplot(projection='3d')
ax.scatter3D(melons['width'],melons['height'],melons['weight'], color = "red")
ax.scatter3D(ballons['width'],ballons['height'],ballons['weight'], color = "orange")

ax.set_xlabel('width')
ax.set_ylabel('height')
ax.set_zlabel('weight')
ax.legend(['melons', 'ballons'])
plt.title("Details")
plt.show()
```
